Design a single unit perceptron for classification of a linearly separable binary dataset (placement.csv) without using pre-defined models. Use the Perceptron() from sklearn.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import Perceptron
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv('Placement.csv')

X = df.iloc[:, 0:2]
y = df.iloc[:, -1]
```

```python
p = Perceptron()
p.fit(X, y)
print(p.coef_)
print(p.intercept_)
print(p.score(X, y))
plot_decision_regions(X.values, y.values, clf=p, legend=2)
```
